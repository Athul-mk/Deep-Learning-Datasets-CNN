```python
input={'tulip':tulip,"rose":rose,"sunflower":sunflower,"dandelion":dandelion,"daisy":daisy}
output={'tulip':0,"rose":1,"sunflower":2,'dandelion':3,"daisy":4}

```

```python
import cv2
import os
import numpy as np
x=[]
y=[]
for flowers in input:
  path_lst=input[flowers]
  for data in path_lst:
    img_arr=cv2.imread(data)
    img_arr_resize=cv2.resize(img_arr,(224,224))
    # print(img_arr_resize.shape)
    img_arr_resize=img_arr_resize/255
    # print(img_arr_resize)
    img_arr_resize=img_arr_resize.reshape(224,224,3,1)
    x.append(img_arr_resize)
    y.append(output[flowers])
    test=np.array(x)
    train=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense,Conv3D,MaxPooling3D,BatchNormalization,Dropout
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
```

```python
model=Sequential()
model.add(Conv3D(filters=10,kernel_size=(2,2,3),input_shape=(224, 224, 3, 1)))
model.add(Conv3D(filters=10,kernel_size=(2,2,1),activation='relu'))
model.add(Flatten())
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=10,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss=SparseCategoricalCrossentropy(),optimizer=Adam(),metrics=['accuracy'])
```

```python
import numpy as np
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
a=model.fit(x=x_train,y=y_train,batch_size=50,epochs=10,validation_data=(x_test,y_test))
```
